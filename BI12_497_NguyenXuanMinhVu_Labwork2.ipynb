```python
#--------------------------#
#1. Histogram
#--------------------------#
plt.figure()
# the histogram of the data
plt.hist(data_flatten, 50, density=True)
plt.ylabel('Probability density')
plt.title(r'Histogram')

from sklearn.neighbors import KernelDensity

kd = KernelDensity(kernel='gaussian', bandwidth=1).fit(data_flatten.reshape(-1,1))
x_axis = np.linspace(np.min(data_flatten), np.max(data_flatten), 200)
log_dens = kd.score_samples(x_axis.reshape(200,1))

plt.plot(x_axis, np.exp(log_dens), color='r', lw=1, linestyle='-', label='gaussian kernel',)
plt.legend()
plt.show()
```

#--------------------------#
#5. Bar plot - a ranking problem
#--------------------------#
Let's imagine that we have 45 students.
In the final exam, we have math and literature score.

The math score is equal to mid_term * 40% + final * 60%.
Same for the literature score.

The final score is the average of the literature score and math score.

Now, we want to plot the score of each student. However, we want to see the score descending (meaning that the student with highest score will stay on top).

Also, we would like to know more about the top 10 students. Therefore, for the first 3 highest overal rank students, highlight the math and literature score with 1,2, and 3. From 4-10, highlight the component score with a star (*). The student ID is displayed in the left hand of the score bar. The use of panda library is prohibited.

Input file:

(each file has 2 columns, the first column is the midterm score, the second column is the final score)

Literature score:

https://drive.google.com/file/d/1ZTUb80rozqXNj3eC8XSzv0Fc3ytsL8pT/view?usp=sharing

Math score:

https://drive.google.com/file/d/1cn_2wTLzGO45adLCSzqbCbQ6I-U8xtgC/view?usp=sharing

image.png

```python
#--------------------------#
#5. Bar plot - a ranking problem
#--------------------------#
txt_ls = open("HW2_literature_score.txt", "r")
list_ls = txt_ls.readlines()
list_ls
ls_score = []
for i in list_ls:
  temp = i.split(" ")
  temp = list(map(float, temp))
  ls_score.append(temp)
ls_score
```

```python
ls_total_score = []
for i in range(len(ls_score)):
  score = ls_score[i][0]*0.4 + ls_score[i][1]*0.6
  ls_total_score.append(score)
ls_total_score = np.array(ls_total_score)
```

```python
sorted_ls_score_index = (-ls_total_score).argsort()
ls_midterm = [ls_score[i][0]*40/100 for i in sorted_ls_score_index]
ls_final = [ls_score[i][1]*60/100 for i in sorted_ls_score_index]
```

```python
txt_ms = open("HW2_math_score.txt", "r")
list_ms = txt_ms.readlines()
list_ms
ms_score = []
for i in list_ms:
  temp = i.split(" ")
  temp = list(map(float, temp))
  ms_score.append(temp)
ms_score
```

```python
ms_total_score = []
for i in range(len(ms_score)):
  score = ms_score[i][0]*0.4 + ms_score[i][1]*0.6
  ms_total_score.append(score)
ms_total_score = np.array(ms_total_score)
ms_total_score
```

```python
sorted_ms_score_index = (-ms_total_score).argsort()
ms_midterm = [ms_score[i][0]*40/100 for i in sorted_ms_score_index]
ms_final = [ms_score[i][1]*60/100 for i in sorted_ms_score_index]
```

```python
# Calculate total score of each student
total_score = [((ms_score[i][0]*40/100 + ms_score[i][1]*60/100)*0.5 + (ls_score[i][0]*40/100 + ls_score[i][1]*60/100)*0.5) for i in range(len(ls_score))]
total_score = np.array(total_score)
total_score
```

```python
# Calculate each score component in comparison with total score (maximum 1)
sorted_total_score_index = (-total_score).argsort()
ls_midterm_by_total = np.array([ls_score[i][0]*0.4*0.5 for i in sorted_total_score_index])
ls_final_by_total = np.array([ls_score[i][1]*0.6*0.5 for i in sorted_total_score_index])
ms_midterm_by_total = np.array([ms_score[i][0]*0.4*0.5 for i in sorted_total_score_index])
ms_final_by_total = np.array([ms_score[i][1]*0.6*0.5 for i in sorted_total_score_index])
ls_midterm_by_total + ls_final_by_total + ms_midterm_by_total + ms_final_by_total
```

```python
#Set up figure which has 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figwidth(12)
fig.set_figheight(8)

# First horizontal bar chart
ax1.title.set_text("(a) Literature Score")
ax1.barh(np.arange(1,len(sorted_ls_score_index)+1), ls_midterm, color = '#fff9e3', height = 0.8)
ax1.barh(np.arange(1,len(sorted_ls_score_index)+1), ls_final, left = ls_midterm, color = '#f3e99d', height = 0.8)
ax1.set_yticks(np.append([1], np.arange(5,len(sorted_ls_score_index)+1, 5)))
ax1.set_xticks(np.arange(0, 1, 0.2))
ax1.set_ylim(0.3, len(sorted_ls_score_index)+0.7)
ax1.grid(axis='x')
ax1.set_axisbelow(True)
for i, index in enumerate(sorted_ls_score_index):
  ax1.text(0.01, i+1.2, index + 1, size = 8)
ax1.invert_yaxis()

# Second horizontal bar chart
ax2.title.set_text("(b) Math Score")
ax2.barh(np.arange(1,len(sorted_ms_score_index)+1), ms_midterm, color = '#e5ffff', height = 0.8)
ax2.barh(np.arange(1,len(sorted_ms_score_index)+1), ms_final, left = ms_midterm, color = '#bbf1f1', height = 0.8)
ax2.set_yticks(np.append([1], np.arange(5,len(sorted_ms_score_index)+1, 5)))
ax2.set_xticks(np.arange(0, 1, 0.2))
ax2.set_ylim(0.3, len(sorted_ms_score_index)+0.7)
ax2.grid(axis='x')
ax2.set_axisbelow(True)
for i, index in enumerate(sorted_ms_score_index):
  ax2.text(0.01, i+1.2, index + 1, size = 8)
ax2.invert_yaxis()

# Third horizontal bar chart
ax3.title.set_text("(c) Final Score")
ax3.barh(np.arange(1,len(sorted_total_score_index)+1), ls_midterm_by_total, color = '#fff9e3', height = 0.8)
ax3.barh(np.arange(1,len(sorted_total_score_index)+1), ls_final_by_total, left = ls_midterm_by_total, color = '#f3e99d', height = 0.8)
ax3.barh(np.arange(1,len(sorted_total_score_index)+1), ms_midterm_by_total, left = ls_final_by_total+ls_midterm_by_total, color = '#e5ffff', height = 0.8)
ax3.barh(np.arange(1,len(sorted_total_score_index)+1), ms_final_by_total, left = ls_final_by_total+ls_midterm_by_total+ms_midterm_by_total, color = '#bbf1f1', height = 0.8)
ax3.set_yticks(np.append([1], np.arange(5,len(sorted_total_score_index)+1, 5)))
ax3.set_xticks(np.arange(0, 0.7, 0.2))
ax3.set_ylim(0.3, len(sorted_total_score_index)+0.7)
ax3.grid(axis='x')
ax3.set_axisbelow(True)
for i, index in enumerate(sorted_total_score_index):
  ax3.text(0.01, i+1.2, index + 1, size = 8)
ax3.invert_yaxis()

# Set order of top 10 students by total score
for i in range(3):
  index1 = np.where(sorted_ls_score_index == sorted_total_score_index[i])[0][0]
  ax1.text(ls_midterm[index1]+ ls_final[index1] + 0.012, index1 + 1.2, i+1, size = 8, weight = 'bold')
  index2 = np.where(sorted_ms_score_index == sorted_total_score_index[i])[0][0]
  ax2.text(ms_midterm[index2]+ ms_final[index2] + 0.012, index2 + 1.2, i+1, size = 8, weight = 'bold')
  ax3.text(ls_midterm_by_total[i] + ls_final_by_total[i] + ms_midterm_by_total[i] + ms_final_by_total[i] + 0.012, i+1.2, i+1, size = 8, weight = 'bold')

for i in range(3, 10):
  index1 = np.where(sorted_ls_score_index == sorted_total_score_index[i])[0][0]
  ax1.text(ls_midterm[index1]+ ls_final[index1] + 0.012, index1 + 1.2, "*", size = 8, weight = 'bold')
  index2 = np.where(sorted_ms_score_index == sorted_total_score_index[i])[0][0]
  ax2.text(ms_midterm[index2]+ ms_final[index2] + 0.012, index2 + 1.2, "*", size = 8, weight = 'bold')
  ax3.text(ls_midterm_by_total[i] + ls_final_by_total[i] + ms_midterm_by_total[i] + ms_final_by_total[i] + 0.012, i+1.2, "*", size = 8, weight = 'bold')

#Show the figure
plt.show()
```

```python
#--------------------------#
#6. Homework 1 without panda
#--------------------------#
sample_data = open("Homework_sampledata1.txt", "r")
list_data = sample_data.readlines()[1:]
data = []
for i in list_data:
  temp = i.split()
  temp = list(map(float, temp))
  data.append(temp)
```

```python
np_data2 = np.array(data)
np_data2 = np_data2[:, 1:]
# -1 means that the dates are not exist
# -99 means that the dates have missing value
for i in [1, 3, 5, 8, 10]:
  if (np_data2[30][i] == -99):
    np_data2[30][i] = -1
if np_data2[29][1] == -99:
   np_data2[29][1] = -1

data_flatten2 = np_data2.T.flatten()

data_flatten2[data_flatten2 == -99] = math.nan
data_flatten2  = data_flatten2[data_flatten2 != -1]

fig = plt.figure()
x = range(0, 366)
ax = fig.add_axes([0.5, 0.1, 1.5, 0.5])
ax.plot(x, data_flatten2, color = 'b')
ax.set_xticks([0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335])
ax.set_xticklabels(["01\n Jan 2008", "01\n Feb", "01\n Mar", "01\n Apr", "01\n May", "01\n Jun", "01\n Jul", "01\n Aug", "01\n Sep", "01\n Oct", "01\n Nov", "01\n Dec"])
plt.xlabel("Time")
plt.ylabel("Temp (oC)")
plt.grid(True, color='gray')
plt.xlim(0, 365)
plt.ylim(5, 35)
plt.legend(["Temperature"])
plt.show()
```

```python
np_data3 = np.array(data)
np_data3 = np_data3[:, 1:]
average = []
for i in range(np_data3.shape[1]):
  sum = 0
  count = 0
  for j in range(np_data3.shape[0]):
    if np_data3[j][i] != -99 and np_data3[j][i] != -1:
      sum += np_data3[j][i]
      count += 1
  average.append(sum/count)

plt.figure(figsize = (12, 3))
plt.plot(average, '', linewidth = 1.5, color = 'b', markersize = 2.5)
plt.xticks(np.arange(12), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.scatter(np.arange(12), average, c= "b", s = 9)
plt.ylim(5, 35)
plt.ylabel("Temp (oC)")
plt.legend(["Temperature"])
plt.show()
```
